```python
path="/home/mw/input/andrew_ml_ex14179/ex1data1.txt"
data=pd.read_csv(path,header=None,names=['Population','Profit'])
data.head()
```

```python
data.plot(kind='scatter',x='Population',y='Profit',figsize=(12,8))
plt.show()
```

```python
#计算J(Ѳ)——损失函数，X为矩阵
def computeCost(X,y,theta):
    inner=np.power(((X*theta.T)-y),2)
    return np.sum(inner)/(2*len(X))
```

```python
#加上Ones列，Ones列的值恒为1,即表示theta0的系数恒为1
data.insert(0,'Ones',1)
#初始化X和y
cols=data.shape[1]
X=data.iloc[:,:-1]
y=data.iloc[:,cols-1:cols]
```

```python
X.head()
```

```python
y.head()
```

```python
X=np.matrix(X.values)
y=np.matrix(y.values)
theta=np.matrix(np.array([0,0]))
```

```python
X.shape,theta.T.shape,y.shape
```

```python
#计算损失函数的值
computeCost(X,y,theta)
```

```python
#梯度下降法对theta进行更新
def gradientDescent(X,y,theta,alpha,iters):
    temp=np.matrix(np.zeros(theta.shape))
    parameters=int(theta.ravel().shape[1])
    cost=np.zeros(iters)
    for i in range(iters):
        error=(X*theta.T)-y
        #注意tenta的同步更新（视频中强调的）
        for j in range(parameters):
            term=np.multiply(error,X[:,j])
            temp[0,j]=theta[0,j]-((alpha/len(X))*np.sum(term))
        theta=temp
        cost[i]=computeCost(X,y,theta)
    return theta,cost
```

```python
#学习率
alpha=0.01
#迭代次数
iters=1500
```

```python
g, cost = gradientDescent(X, y, theta, alpha, iters)
g
```

```python
cost
```

```python
predict1=[1,3.5]*g.T
print("predict1:",predict1)
predict2=[1,7]*g.T
print("predict2",predict2)
```

```python
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False,dtype=None)产生从start到stop的等差数列，num为元素个数，默认50个
x = np.linspace(data.Population.min(),data.Population.max(),100)
f=g[0,0]+(g[0,1]*x)
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(x,f,'r',label='Prediction')
ax.scatter(data.Population,data.Profit,label='Traing Data')
ax.legend(loc=2)
ax.set_xlabel('Population')
ax.set_ylabel('Profit')
ax.set_title('Predicted Profit vs. Population Size')
plt.show()
```
